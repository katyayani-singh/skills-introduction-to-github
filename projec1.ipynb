```python
# Simple Restaurant Ordering Chatbot

def display_menu():
    menu = {
        "pizza": 129,
        "burger": 99,
        "pasta": 119,
        "noodles": 119,
        "drink": 89
    }
    print("\nMenu:")
    for item, price in menu.items():
        print(f"{item.capitalize()} - ${price:.2f}")
    return menu

def chatbot():
    print("Welcome to our restaurant! Type 'exit' to quit.")
    menu = display_menu()
    order = {}

    while True:
        user_input = input("\nWhat would you like to order? ").lower()

        if user_input == "exit":
            break
        elif user_input in menu:
            quantity = int(input(f"How many {user_input}s would you like? "))
            if user_input in order:
                order[user_input] += quantity
            else:
                order[user_input] = quantity
            print(f"{quantity} {user_input}(s) added to your order.")
        else:
            print("Sorry, we don't have that item.")

    if order:
        total = sum(menu[item] * qty for item, qty in order.items())
        print("\nYour Order:")
        for item, qty in order.items():
            print(f"{item.capitalize()} x{qty} - ${menu[item] * qty:.2f}")
        print(f"\nTotal Bill: ${total:.2f}")
    else:
        print("No items ordered. Thank you for visiting!")

    print("Have a great day!")

# Run the chatbot
chatbot()
```

```python
#simple library assistant chatbot
def library_chatbot():
    print("Welcome to the Library Chatbot!")
    print("How can I help you today?")

    while True:
        query = input(">").lower()  # Convert input to lowercase for easier matching

        if "hours" in query:
            print("The library is open Monday-Friday, 9am to 5pm.")

        elif "book" in query:
            book_title = input("Please enter the title of the book: ").lower()
            #  In a real implementation, you would connect to a library database
            #  and search for the book.  For this example, we have a static list.

            book_data = {
                "the lord of the rings": {
                    "author": "J.R.R. Tolkien",
                    "availability": "Checked out (due back 10/27)",
                },
                "pride and prejudice": {
                    "author": "Jane Austen",
                    "availability": "Available",
                },
                "1984": {
                    "author": "George Orwell",
                    "availability": "Available",
                },
                "the hobbit": {
                    "author": "J.R.R. Tolkien",
                    "availability": "Available",
                },
                "sense and sensibility": {
                    "author": "Jane Austen",
                    "availability": "Available",
                },
                "animal farm": {
                    "author": "George Orwell",
                    "availability": "Available",
                },
                "the great gatsby": {
                    "author": "F. Scott Fitzgerald",
                    "availability": "Available",
                },
                "the adventures of huckleberry finn": {
                    "author": "Mark Twain",
                    "availability": "Available",
                },
                "the adventures of sherlock holmes": {
                    "author": "Arthur Conan Doyle",
                    "availability": "Available",
                },
                "the adventures of tom sawyer": {
                    "author": "Mark Twain",
                    "availability": "Available",
                },
                "the adventures of alice in wonderland": {
                    "author": "Lewis Carroll",
                    "availability": "Available",
                }
            }

            if book_title in book_data:
                print(f"Title: {book_title.title()}") # Capitalize first letter
                print(f"Author: {book_data[book_title]['author']}")
                print(f"Availability: {book_data[book_title]['availability']}")
            else:
                print(f"Sorry, we don't have information on the book '{book_title}' in our catalog.")

        elif "author" in query:
            author_name = input("Please enter the author's name: ").lower()

            #  Again, this would ideally connect to a database.
            author_books = {
                "j.r.r. tolkien": ["The Lord of the Rings", "The Hobbit"],
                "jane austen": ["Pride and Prejudice", "Sense and Sensibility"],
                "george orwell": ["1984", "Animal Farm"]
            }

            if author_name in author_books:
                print(f"Books by {author_name.title()}:")
                for book in author_books[author_name]:
                    print(f"- {book}")
            else:
                print(f"Sorry, we don't have any books listed under the author '{author_name}'.")

        elif "hello" in query or "hi" in query:
            print("Hello! How can I assist you today?")

        elif "thank you" in query or "thanks" in query:
            print("You're welcome!")

        elif "exit" in query or "quit" in query or "bye" in query:
            print("Goodbye!")
            break  # Exit the loop and end the chatbot session

        else:
            print("I'm sorry, I don't understand.  I can help with information about library hours, books, and authors.")


if __name__ == "__main__":
    library_chatbot()
```
